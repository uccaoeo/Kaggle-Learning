```python
import autogluon as ag
from autogluon.tabular import TabularPredictor as task
from autogluon.tabular import TabularDataset
```

```python
import pandas as pd
import numpy as np
from google.colab import files
import io
```

```python
uploaded = files.upload()
```

```python
df= pd.read_csv(io.BytesIO(uploaded['train.csv']))
```

```python
train = TabularDataset(df)
train.head()
```

```python
predictor = task.load("AutogluonModels/ag-20210519_192139/")
```

```python
label='SalePrice'
```

```python

predictor = task(label=label).fit(train)  # presets='high_quality_fast_inference_only_refit'  time_limits=60*60*8
```

```python
uploaded = files.upload()
```

```python
testdf= pd.read_csv(io.BytesIO(uploaded['test.csv']))
```

```python
test = TabularDataset(testdf)
test.head()
```

```python
y_pred = predictor.predict(test)
```

```python
predictor.evaluate(train)
```

```python
uploaded = files.upload()
```

```python
submi= pd.read_csv(io.BytesIO(uploaded['sample_submission.csv']))
```

```python
import pandas as pd
submi['SalePrice'] = y_pred
```

```python
submi.drop(['Survived'], axis=1, inplace=True)
```

```python
submi.to_csv('submissionautogluon.csv', index=False)
submi.head()
files.download('submissionautogluon.csv')
```
